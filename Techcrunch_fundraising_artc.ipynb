```python
import bs4
import pandas as pd
import requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
url = 'https://techcrunch.com/techcrunchplus/fundraising/'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
html = response.content
pprint(html)
```

```python
article_titles, article_contents, article_authors, article_hrefs = [], [], [], []

for tag in soup.findAll("div", {"class": "post-block post-block--image post-block--unread"}):
    tag_header = tag.find("a", {"class": "post-block__title__link"})
    tag_author = tag.find("span", {"class": "river-byline__authors"})
    tag_time = tag.find("time", {"class": "river-byline__full-date-time"})
    tag_content = tag.find("div", {"class": "post-block__content"})

    article_title = tag_header.get_text().strip()
    article_author = tag_author.get_text().strip()
    article_content = tag_content.get_text().strip()
    article_href = tag_header["href"]

    article_titles.append(article_title)
    article_authors.append(article_author)
    article_contents.append(article_content)
    article_hrefs.append(article_href)
```

```python
df = pd.DataFrame({"title": article_titles, "content": article_contents, "author": article_authors, "link": article_hrefs})
df.shape
```
